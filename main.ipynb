```python
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})
```

```python
y = df["diagnosis"]

x = df.drop(columns=["diagnosis", "id"], axis=1)

```

```python
from LogisticRigression import LogisticRegression, accuracy,train_test_split,zscore_normalize_features
```

```python
x = zscore_normalize_features(x)
```

```python
x
```

```python
X_train, y_train, X_test, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
logistic = LogisticRegression(learning_rate=0.01,iteration=2500)
logistic.fit(X_train, y_train)
predictions = logistic.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(logistic.error)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

sns.heatmap(data = confusion_matrix(y_test, predictions) , annot = True , linewidth = 0.1 , cmap='Reds')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, predictions))

model_accuracy = accuracy(y_test, predictions)
print(f"Accuracy of the Logistic Regression Model is: {model_accuracy * 100 : .2f} %")
```
